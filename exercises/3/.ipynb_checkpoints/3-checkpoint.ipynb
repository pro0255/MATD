```python
PATTERN = 'AGAGACAAGATTCCA'
PATTERN1 = 'TACTACATCTGACGCACAATCCC'
```

```python
def create_bad_match_table(pat):
    res = {}
    m = len(pat)
    print(pat)
    for i in range(m-1):
        pos = m - i - 1
        character = pat[i]
        res[character] = pos
    return res
```

```python
def BMH_locate(pat, txt):
    positions = []
    
    table = create_bad_match_table(pat)
    m = len(pat)
    n = len(txt)
    i = 0
    
    #print(table)
    
    while(i < n-m):
        current_txt = txt[i:i+m]
        j = m - 1

        while(current_txt[j] == pat[j]):
            if j == 0:
                break
            j -= 1
        
        if j == 0:
            positions.append((i, i+m))
            #print(f'Founded at {i}')
            i += 1
        else:
            bad_character = current_txt[j]
            shift = table.get(bad_character, m)
            #print(f'Shifted with {shift}')
            i += shift
                
    return positions 
```
