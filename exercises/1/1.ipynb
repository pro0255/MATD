```python
def resolve_alphabet(doc):
    return np.unique(list(doc))
```

```python
#state 0 is empty word

def build_table(pattern):
    #vrati dictionary[state][symbol]
    pattern_l = len(pattern)
    
    alphabet = resolve_alphabet(pattern)
    table = {i:{char:0 for char in alphabet} for i in range(pattern_l+1)}
    
    state2Prefix = {}
    
    for i in range(pattern_l+1):
        if i == 0:
            state2Prefix[0] = ''
            continue
        state2Prefix[i] = pattern[0:i]
        
        
    for i in range(pattern_l+1):
        current = ''
        
        if i != 0:
            current = pattern[0:i]

        new_states = [current+char for char in alphabet]
        
        for state in new_states:
            is_prefix = pattern.startswith(state)
            if is_prefix:
                last_char = state[-1]
                new_states = list(filter(lambda x: x != state,new_states))
                table[i][last_char] = len(state)
        
        
        for a in new_states:
            last_char = a[-1]
            move_state = 0
            
            prefixes = list(filter(lambda x: len(x) != len(a), [a[-j:] for j in range(len(a))]))
            sorted_prefixes = sorted(prefixes, key=len, reverse=True)
            
            for prefix in sorted_prefixes:
                if pattern.startswith(prefix):
                    move_state = len(prefix) 
                    break
            table[i][last_char] = move_state
            
    return (table, alphabet)
```

```python
def locate(pattern, text):
    table, alphabet = build_table(pattern)
    state = 0
    
    print(table, alphabet)
    
    positions = []
    
    for position, c in enumerate(text):
        if c not in alphabet:
            state = 0
            continue
            
        state = table[state][c]        
        is_founded = state == len(table) - 1
        
        if is_founded:
            
            end = position
            start = end - len(pattern) + 1
            positions.append((start, end))
            
    #vrati vyskyty patternu v textu
    return positions
```
